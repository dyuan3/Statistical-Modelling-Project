# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests

data = requests.get("http://api.citybik.es/v2/networks/velib").json() 
```

```python
data
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
data_dict = {

    'Lat' : [],
    'Long':[],
    'No bikes':[]
}

for i in range(len(data['network']['stations'])):
  data_dict['Lat'].append(data['network']['stations'][i]['latitude'])
  data_dict['Long'].append(data['network']['stations'][i]['longitude'])
  data_dict['No bikes'].append(data['network']['stations'][i]['free_bikes'])
```

```python
data_dict['Lat']
```

Put your parsed results into a DataFrame.

```python
import pandas as pd
pd.DataFrame(data_dict)
```
